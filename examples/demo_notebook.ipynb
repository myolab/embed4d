This notebook shows how to display 3D GLB/GLTF/FBX models inline using `notebook_viewer`.

**Requirements:** `pip install embed4d[ipython]` (or `pip install ipython`)

## Empty viewer (drag & drop)

Display an empty viewer; you can drag and drop a `.glb` / `.gltf` / `.fbx` file into it.

```python
from embed4d import notebook_viewer

notebook_viewer()
```

## Load a model from a file path

Pass a path to a GLB/GLTF/FBX file to embed it in the viewer.

```python
from embed4d import notebook_viewer

# Replace with your model path
notebook_viewer("path/to/your/model.glb", height=600)
```

## Custom height

You can set a custom height (in pixels) for the viewer.

```python
from embed4d import notebook_viewer

notebook_viewer(height=800)
```
